```python
# one max
NUM_BITS=100
POP_SIZE=10
OFFSPRING_SIZE=5
```

```python
import random
```

```python
population=[ random.choices([0,1],k=NUM_BITS) for i in range(POP_SIZE)]
```

```python
def fitness(ind):
    return sum(ind)
```

```python
def select(pop):
    # tournament  selction
    ind1, ind2 = random.choices(range(len(pop)), k=2)
    if fitness(pop[ind1]) > fitness(pop[ind2]):
        return ind1
    else:
        return ind2
```

```python
for q in range(100):
    offspring=[]
    for _ in range(OFFSPRING_SIZE):
        p=select(population)
        new_ind=population[p][:]
        bit =random.randint(0,len(new_ind)-1)
        if new_ind[bit]==0:
            new_ind[bit]=1
        else:
            new_ind[bit]=1
        offspring.append(new_ind)
    population+=offspring
    population=sorted(population,key=lambda i:fitness(i),reverse=True)
    population=population[:POP_SIZE]
    print(f"max fitness {fitness(population[0])}")
```
